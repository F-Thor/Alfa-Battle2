```python
def max_pow (x):
    i=1
    while 2**i<x:
        i+=1
    return(i)
```

```python
def dt_feath_2_vec(feathers):
    for feather in feathers:
        cat=(pd.read_csv(feather+'.csv')).values
        categories=[]
        for i in cat:    
            categories.extend(i)
        pow=max_pow(len(categories))
        categories.sort()
        a=[]
        for i in product([0,1],repeat=pow):
            a.append(i)
        n=np.array(a)
        e=np.reshape(n, (-1,pow))
        z=np.random.randint(0,(len(categories)),(2**pow-len(categories)))
        for i in z:
            e = np.delete(e, (i), axis=0)
        v_cat=pd.DataFrame(e,index=categories)
        v_cat.to_csv((feather+'_vec'+'.csv'))
        print(v_cat)
```

```python
def feath_2_vec(feathers):
    for feather in feathers:
        cat_O=pd.read_csv(feather+'.csv')#.values
        cat_O['0'].fillna('-1', inplace=True)
        #print(cat_O.iloc[0,0])
        #cat_O.iloc[2]='-1' 
        cat=cat_O.values
        categories=[]
        for i in cat:    
            categories.extend(i)
        pow=max_pow(len(categories))
        a=[]
        for i in product([0,1],repeat=pow):
            a.append(i)
        n=np.array(a)
        e=np.reshape(n, (-1,pow))
        z=np.random.randint(0,(len(categories)),(2**pow-len(categories)))
        for i in z:
            e = np.delete(e, (i), axis=0)
        v_cat=pd.DataFrame(e,index=categories)
        v_cat.to_csv((feather+'_vec'+'.csv'))
        print(feather,v_cat,(cat_O['0'].isnull()).value_counts())
```
